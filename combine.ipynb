```python
import pandas as pd
from pathlib import Path 
from tqdm import tqdm
```

```python
OUTPUT_FOLDER = Path("output")
```

```python
dataframes = []
filenames = list(OUTPUT_FOLDER.glob("**/results.pkl"))
filenames.sort()
for session_name in tqdm(filenames):
    df = pd.read_pickle(session_name)
    new_index = [(session_name.parent.name, x.split('-')[0]) for x in df.index]
    df.index = pd.MultiIndex.from_tuples(new_index)
    dataframes.append(df)

dataframes = pd.concat(dataframes)
dataframes
```

```python
dataframes.to_excel("output/combined-results.xlsx")
```
